Linking of probability and QALM:


QALM value at Node A =
(Probability at Node B *  QALM at Node B) + (Probability at Node C * QALM at Node C)

QALM at terminal Terminal Node =
(Utility of the survival state * No. of months patient  stayed in that month) + Observed QALM
